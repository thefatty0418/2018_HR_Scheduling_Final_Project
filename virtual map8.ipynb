```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
import os
import chardet
import time
```

```python
# 設定檔案路徑

dir_data = './data'
```

```python
# 讀入所有車站、停車場、商圈資訊
f_loc = os.path.join(dir_data, 'loc.csv')
print('path of read in data: %s' % (f_loc))


with open(f_loc, 'rb') as f:
    result = chardet.detect(f.read())
    
f_loc_data = pd.read_csv(f_loc,encoding=result['encoding'])

f_loc_data.head()
```

```python
# 讀入商圈位置資訊
com_loc = os.path.join(dir_data, 'Commercial_district.csv')
print('path of read in data: %s' % (com_loc))


with open(com_loc, 'rb') as f:
    result = chardet.detect(f.read())
    
com_loc_data = pd.read_csv(com_loc,encoding=result['encoding'])

com_loc_data.head()
```

```python
# 整體台北商圈人數統計值
com_loc_data['人數'].describe()
```

```python
# 整體台北商圈車位數統計值
com_loc_data['車位數'].describe()
```

```python
#檢視 X Y 座落範圍決定隨機生成潛在商圈的範圍
f_loc_data['X'].describe()
```

```python
f_loc_data['Y'].describe()
```

```python
pd.set_option('colheader_justify', 'left')
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)
from IPython.display import display
# 計算距離的函式
def distance(a):
        currentDistance = 0
        currentDistance = abs(x[i] - a[0]) + abs(y[i] - b[0])
        return (currentDistance)
com_peo_mean = com_loc_data['人數'].mean()
com_park_mean = com_loc_data['車位數'].mean()
count = 0
loc_s = []
# 選出5個台北的潛在商圈位置
start = time.time()
while count < 5:

    #建立初始地圖
    plt.figure(figsize = (20, 16))
    plt.scatter('X','Y',c='color', s='size', data = f_loc_data, alpha= 0.7)
    plt.title('Maps')
    a = []
    b = []

    # 隨機座標點位置為整體座標最大最小範圍 +/- 10
    a.append(int(rd.uniform(-69.6,137.5)))
    b.append(int(rd.uniform(-105.9,144)))
    i = a[0]
    j = b[0]
    if a[0] in com_loc_data['X']:
        i = int(rd.uniform(-69.6,137.5))
    plt.scatter(i, j, s=50, c='m')
    plt.annotate('(%d,%d)' %(a[0],b[0]), xy=(a[0],b[0]), xytext=(a[0]+1,b[0]+1))

    # 建立可比較座標點
    x = []
    y = []
    for i in f_loc_data['X']:
        x.append(i)
    for j in f_loc_data['Y']:
        y.append(j)

    # 判別隨機商圈裡有多少人 並將之以紅色標示
    station_list = []
    for i in range(len(x)):
        if distance(a) <= 13:
            plt.scatter(x[i],y[i], s=50 , c='r')
            """
            因 df.loc的功能無法同時使用兩個數值找出列，因此為避免單純使用 X座標導致取出同樣 Y軸上不正確的點
            因此使用已下雙重定位的方式
            p.s. csv檔案中的數字格式必須要是文字形式才有辦法將找到的座標點 x[i]搭配回去搜尋
            """
            station_x = f_loc_data.loc[lambda f_loc_data: f_loc_data['X'] == x[i] ]
            station = station_x.loc[lambda station_x: station_x['Y'] == y[i] ]
            station_list.append(station)
            # pandas 範例 df.loc[lambda df: df['shield'] == 8]

    station_merge = f_loc_data.head(0) 
    # f_loc_data.head(0)的目的在於提供一個空白的 dataframe供新截取的資料融合
    if len(station_list) > 0:
        left = f_loc_data.head(0)
        for station in station_list:
            right = station
            station_merge = pd.merge(left, right, how='outer')
            # 'outer'的方法是將欄位標題相同的資料merge在一起的語法
            left = station_merge
            
    #比較所選出點與目標商圈的差距大於既有商圈的負0.1即可視為潛在商圈
    station_peo_sum = station_merge['人數'].sum()
    station_park_sum = station_merge['車位數'].sum()
    d_com_stat_people = (station_peo_sum -com_peo_mean) / com_peo_mean
    d_com_stat_park = (station_park_sum -com_park_mean) / com_park_mean
    if d_com_stat_people > -0.1:
        print('此位置有足夠人潮可為潛在商圈')
        print('此位置人潮與現有商圈平均比較: ', '{0:4.2%}'.format(d_com_stat_people))
        print('X:',a[0],'Y:',b[0])
        display(station_merge.head(10)) # 使dataframe不會像文字檔一般被字串長度影響而排版凌亂
        count += 1
        if d_com_stat_park > -0.1:
            print('此位置停車位與現有商圈平均比較: ', '{0:4.2%}'.format(d_com_stat_park))
            print('此位置人潮多且停車方便\n')
print('以上為台北市內潛在商圈')
end = time.time()
elapsed = end - start
print('所花時間: ', elapsed, '秒')    
plt.show()    




```
